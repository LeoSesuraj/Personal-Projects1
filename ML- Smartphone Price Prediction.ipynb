```python
import pandas
from matplotlib import pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
#Load the data
filepath = "/Users/admin/Library/CloudStorage/OneDrive-purdue.edu/Data Sets/Cellphone.csv"
dataframe = pandas.read_csv(filepath)
dataframe.head()
df_clean=dataframe.drop(columns=["Unnamed: 14","Unnamed: 15", "Unnamed: 16", "Unnamed: 17","Unnamed: 18","Unnamed: 19"]) #dropping unwanted columns
df_clean.isnull().sum() #returns how many null values in each column

X=df_clean.drop(columns=["Price"]) #Feature matrix
Y=df_clean["Price"] #Just target
```

```python
plt.figure(figsize=(12,8)) #indicated we are making a figure that is 12in x 8in
heatmap_data= df_clean.corr() #finds the correlation between all data
sns.heatmap(heatmap_data, annot=True, cmap="coolwarm", linewidths=0.5) #making the heatmap by passing it the heatmap data, assigning 
#the values to each square, making the color blue for negative values and red for positive, making sure the square can be seen so linewidth of 0.5
plt.title("Correlation Matrix Heatmap")
```

```python
# Split the data into training and test sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.04, random_state=42)

# Create and train the Random Forest Regressor (no display will occur)
r_forest = RandomForestRegressor(random_state=42);
r_forest.fit(X_train, Y_train)  # Train the model

print("Training complete!")
```

```python
Y_values= r_forest.predict(X_test)

#display results
mae = mean_absolute_error(Y_test, Y_values)
mse = mean_squared_error(Y_test, Y_values)
r2score = r2_score(Y_test, Y_values)

# Calculate percentage accuracy based on R score
accuracy = r2score * 100

print(f"Mean Absolute Error: {mae}")
print(f"Mean Squared Error: {mse}")
print(f"R Score: {r2score}")
print(f"Accuracy: {accuracy:.3f}%")

```
